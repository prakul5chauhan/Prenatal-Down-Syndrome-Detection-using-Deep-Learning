```python
import cv2
import numpy as np
import os

# Define paths
label_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Output/labels"  # Folder containing .txt files
image_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Output"  # Folder with corresponding images
output_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Masks"  # Where masks will be saved
os.makedirs(output_dir, exist_ok=True)

# Get all label files
label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

for label_file in label_files:
    # Read image to get dimensions
    image_name = label_file.replace('.txt', '.jpg')  # or .png
    image_path = os.path.join(image_dir, image_name)
    image = cv2.imread(image_path)
    h, w = image.shape[:2]  # Height & width

    # Initialize blank mask
    mask = np.zeros((h, w), dtype=np.uint8)

    # Read label file
    with open(os.path.join(label_dir, label_file), 'r') as f:
        lines = f.readlines()

    for line in lines:
        parts = list(map(float, line.strip().split()))
        class_id = int(parts[0])
        polygon = parts[1:]  # Normalized coordinates [x1, y1, x2, y2, ...]

        # Convert normalized coords to absolute pixel values
        polygon = np.array(polygon, dtype=np.float32).reshape(-1, 2)
        polygon[:, 0] *= w  # Scale x-coordinates
        polygon[:, 1] *= h  # Scale y-coordinates
        polygon = polygon.astype(np.int32)

        # Draw filled polygon on mask
        cv2.fillPoly(mask, [polygon], color=255)  # 255 = white (foreground)

    # Save binary mask
    mask_filename = os.path.join(output_dir, f"mask_{image_name}")
    cv2.imwrite(mask_filename, mask)
```

```python
import cv2
import numpy as np
import os
from PIL import Image
import csv

# Paths
mask_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Masks"
output_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Measurements"
os.makedirs(output_dir, exist_ok=True)

# Pixel-to-mm conversion from first script
PIXEL_TO_MM = 0.05  # 1 pixel = 0.05 mm

# Optional: CSV logging
csv_path = os.path.join(output_dir, "nt_measurements.csv")
csv_file = open(csv_path, mode='w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Filename", "NT Length (pixels)", "NT Length (mm)", "Risk"])

# Process each mask
for mask_file in os.listdir(mask_dir):
    if mask_file.startswith("mask_") and mask_file.endswith(".jpg"):
        # Load mask in grayscale using PIL
        mask_path = os.path.join(mask_dir, mask_file)
        image = Image.open(mask_path).convert("L")
        image_array = np.array(image)

        # Threshold to binary mask
        binary_mask = image_array > 128
        rows_with_mask = np.any(binary_mask, axis=1)

        if not np.any(rows_with_mask):
            print(f"{mask_file}: No NT region found!")
            continue

        # Measure vertical extent
        top_pixel = np.argmax(rows_with_mask)
        bottom_pixel = len(rows_with_mask) - np.argmax(rows_with_mask[::-1]) - 1
        vertical_length_pixels = bottom_pixel - top_pixel + 1
        vertical_length_mm = vertical_length_pixels * PIXEL_TO_MM

        # Risk classification
        risk = "High Risk" if vertical_length_mm >= 3.5 else "Normal"

        # Annotate result
        mask_bgr = cv2.cvtColor(image_array, cv2.COLOR_GRAY2BGR)
        cv2.line(mask_bgr, (0, top_pixel), (mask_bgr.shape[1], top_pixel), (255, 0, 0), 1)
        cv2.line(mask_bgr, (0, bottom_pixel), (mask_bgr.shape[1], bottom_pixel), (255, 0, 0), 1)
        cv2.putText(mask_bgr, f"NT: {vertical_length_mm:.2f}mm ({risk})", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # Save image
        output_path = os.path.join(output_dir, f"measured_{mask_file}")
        cv2.imwrite(output_path, mask_bgr)

        # Print and log
        print(f"{mask_file}: NT Length = {vertical_length_mm:.2f} mm ({risk})")
        csv_writer.writerow([mask_file, vertical_length_pixels, f"{vertical_length_mm:.2f}", risk])

# Close CSV
csv_file.close()
```

```python
import gradio as gr
import cv2
import numpy as np
import os
from ultralytics import YOLO

# Initialize paths (same as your project)
output_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Measurements"
mask_dir = "/content/drive/MyDrive/Colab Notebooks/PrenatalDownSyndromeDetection/NT Masks"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(mask_dir, exist_ok=True)

# Load trained model
model = YOLO("/content/runs/segment/train/weights/best.pt")

def process_ultrasound(img, pixel_to_mm=0.1):
    """Full pipeline with mask saving"""
    # YOLOv11 Prediction
    results = model.predict(source=img, save=False, imgsz=640)

    if not results or not results[0].masks:
        return img, None, "No NT region detected"

    # Get and save mask (your original logic)
    mask = results[0].masks[0].data[0].cpu().numpy()
    mask = (mask * 255).astype(np.uint8)

    # Generate unique filename
    mask_filename = f"mask_{len(os.listdir(mask_dir))}.png"
    cv2.imwrite(os.path.join(mask_dir, mask_filename), mask)

    # NT Measurement
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    largest_contour = max(contours, key=cv2.contourArea)
    rect = cv2.minAreaRect(largest_contour)
    (x, y), (w, h), _ = rect
    nt_length = max(w, h) * pixel_to_mm
    risk = "High Risk ğŸš¨" if nt_length >= 3.5 else "Normal âœ…"

    # Visualization
    vis = img.copy()
    box = cv2.boxPoints(rect).astype(int)
    cv2.drawContours(vis, [box], 0, (0,255,0), 2)
    cv2.putText(vis, f"{nt_length:.2f}mm ({risk})", (10,30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)

    # Save measurement result (your original code)
    result_filename = f"measured_{mask_filename}"
    cv2.imwrite(os.path.join(output_dir, result_filename), vis)

    return vis, mask, f"NT Length: {nt_length:.2f}mm | Status: {risk}"

# Rest of Gradio interface remains the same
```

```python
import gradio as gr

demo = gr.Interface(
    fn=process_ultrasound,
    inputs=[
        gr.Image(type="numpy", label="Upload Ultrasound Image"),
        gr.Number(value=0.05, label="Pixel-to-mm Ratio")
    ],
    outputs=[
        gr.Image(label="Annotated Image"),
        gr.Image(label="NT Mask"),
        gr.Textbox(label="Measurement Result")
    ],
    title="Prenatal NT Measurement",
    description="Upload a fetal ultrasound image to measure NT length and assess Down Syndrome risk."
)

demo.launch(share=True, debug=True)
```
